```python
import pandas as pd
import mplfinance as mpf
import yfinance as yf
```

```python
dados = yf.download(tickers='LJQQ3.SA', start='2023-01-01', end='2023-12-31')
```

```python
s = mpf.make_mpf_style(base_mpf_style='yahoo',legacy_rc={'figure.titlesize':30})

mpf.plot(dados.head(60), type='candle', figsize=(16,8), volume=True, 
        xlabel='Perodo',ylabel = 'Valor das Aes',ylabel_lower='Volume',
        mav=(7,14), style=s,title='LJQQ3.SA - Lojas Quero Quero',
        datetime_format=' %d/%m/%Y',xrotation=20)
```
